Decimos que una serie de tiempo con una variacin constante es **estacionaria**. Las propiedades de una serie temporal estacionaria no varan con el tiempo.

Existe un patrn **estacional** cuando una serie est influenciada por factores estacionales (por ejemplo, el trimestre del ao, el mes o el da de la semana). La estacionalidad es siempre *de un perodo fijo y conocido*. Por lo tanto, las series temporales estacionales a veces se denominan **peridicas** series temporales.

Existe un patrn **cclico** cuando los datos muestran subidas y cadas que *no son de perodo fijo*. Por ejemplo, los ciclos econmicos que suelen durar varios aos, pero de antemano se desconoce la duracin del ciclo actual.

Existe una **tendencia** cuando hay un aumento o disminucin a largo plazo en los datos. No tiene que ser lineal. A veces nos referiremos a una tendencia que "cambia de direccin" cuando podra ir de una tendencia creciente a una tendencia decreciente.

En el grfico anterior, notamos las siguientes propiedades de esta serie de tiempo.

- Esta serie temporal tiene una tendencia significativa.
- Tiene un componente estacional.
- La duracin del ciclo (en meses) vara con el tiempo.
- El grfico de la serie logartmica transformada parece tener un tamao ligeramente ms uniforme para el componente estacional que el no transformado. As que vamos a utilizar la serie de log transformado para un anlisis ms detallado

Cualquier serie temporal con una tendencia implica que vara con el tiempo, lo que significa que no es estacionario**.

Una serie temporal estacionaria sin patrn estacional o cclico y sin tendencia perceptible se dice que es **independiente distribuida de manera idntica** o **i.i.d.** An i.i.d. Las series de tiempo tienen una distribucin normal con una distribucin en forma de campana y una grfica Q-Q de lnea recta. Tal serie de tiempo tambin se llama ruido blanco.

Vamos a trazar la distribucin de nuestras series de tiempo a continuacin:

```r
# Grfico la distribucin - histograma y QQ de la serie temporal
dist.ts = function(df, col = 'residual', bins = 40)
{
  par(mfrow = c(1,2))
  temp = as.vector(df)
  breaks = seq(min(temp), max(temp), length.out = (bins + 1))
  hist(temp, breaks = breaks, main = paste('Distribution of', col), xlab = col)
  qqnorm(temp, main = paste('Q-Q Normal plot of', col))
  par(mfrow = c(1,1))
}

options(repr.pmales.extlot.width=8, repr.plot.height=4)
dist.ts(log(itca.pm25.ts) , col = 'PM2.5 Itca')
dist.ts(log(itca.pm10.ts), col = 'PM10 Itca')
dist.ts(log(itca.temp.ts), col = 'Temperatura Itca')
```

La grfica Q-Q no es una lnea recta y, por lo tanto, esto no es un i.i.d. distribucin normal. Esto confirma nuestros hallazgos iniciales de que este conjunto de datos tiene ciertos patrones y tendencias inherentes.

A continuacin exploremos algunas formas de detectar patrones.

**Autocorrelacin**, tambin conocida como correlacin en serie, es la correlacin de una serie de tiempo con una copia retrasada de s misma. Informalmente, es la similitud entre las observaciones en funcin del tiempo de retraso entre ellas. El anlisis de la autocorrelacin es una herramienta matemtica para encontrar patrones repetitivos, como la presencia de periodicidad o estacionalidad.

En estadstica, la autocorrelacin de un proceso aleatorio es la correlacin de Pearson entre los valores del proceso en diferentes momentos, en funcin de los dos tiempos o del retraso. Esto se denomina funcin de correlacin automtica o ACF. El ACF es una de las herramientas utilizadas para encontrar patrones en los datos. Especficamente, el ACF le indica la correlacin entre puntos separados por varios retrasos de tiempo. Tenga en cuenta que para cualquier serie, la autocorrleacin en el retardo cero es igual a uno.

Tambin podemos definir una segunda orden **autocorrelacin parcial**. La funcin de autocorrelacin parcial (PACF) proporciona la correlacin parcial de una serie de tiempo con sus propios valores rezagados, controlando los valores de la serie de tiempo en todos los rezagos ms cortos. Contrasta con la funcin de autocorrelacin, que no controla para otros retrasos. Esta funcin desempea un papel importante en los anlisis de datos dirigidos a identificar la extensin del retraso en un modelo autorregresivo (AR), que describiremos en breve.

Ahora, vamos a trazar el ACF y PACF de nuestros datos de tasa de desempleo

```r
#Funciones de autocorrelacin y autocorrelacin parcial
plot.acf <- function(df, col = 'remainder', is.df =TRUE)
{
  if(is.df) temp <- df[, col]
  else temp <- df
  par(mfrow = c(2,1))
  acf(temp, main = paste('ACF de', col))
  pacf(temp, main = paste('PACF de', col))
  par(mfrow = c(1,1))
}
      
options(repr.pmales.extlot.width=8, repr.plot.height=6)
plot.acf(log(itca.pm25.ts), col = 'Tasa de Material Particulado 2.5', is.df = F)
plot.acf(log(itca.pm10.ts), col = 'Tasa de Material Particulado 10', is.df = F)
plot.acf(log(itca.temp.ts), col = 'Tasa de Material Temperatura', is.df = F)
```

Las lneas de puntos azules en los grficos ACF y PACF son los intervalos de confianza del 95% para los valores de correlacin. Las lneas verticales muestran la correlacin. Cualquier lnea vertical que cruce las lneas punteadas azules indica un valor de correlacin significativo. Un i.i.d. las series de tiempo normalmente distribuidas no tendrn valores significativos de ACF o PACF ms all del retraso 0. En este caso, el grfico de ACF tiene varios retrasos significativos no cero y muestra cierto patrn de aumento y cada. Tambin el ACF est disminuyendo o decayendo lentamente. La grfica PACF muestra valores significativos para ms de un retraso, lo que indica que podra haber una tendencia. Esto claramente no es una serie temporal estacionaria.

El proceso de extraccin de la tendencia, el ciclo y los componentes estacionales se denomina descomposicin. A continuacin, veremos los mtodos de descomposicin.

**Descomposicin basada en tasas de cambio**

Esta es una tcnica importante para todos los tipos de anlisis de series de tiempo, especialmente para el ajuste estacional. Busca construir, a partir de una serie temporal observada, una serie de componentes (que podran usarse para reconstruir el original mediante adiciones o multiplicaciones) donde cada una de ellas tiene una cierta caracterstica o tipo de comportamiento. Por ejemplo, las series de tiempo generalmente se descomponen en:

$ T_t $, el componente de tendencia en el tiempo t, que refleja la progresin a largo plazo de la serie.

$ C_t $, el componente cclico en el momento t, que describe fluctuaciones repetidas pero no peridicas.

$ S_t $, el componente estacional en el tiempo t, que refleja la estacionalidad.

$ I_t $, el componente irregular (o "ruido") en el tiempo t, que describe influencias aleatorias e irregulares. Esta es la parte de la serie que no se puede atribuir a los componentes estacionales, de ciclo o de tendencia y se conoce como residual, resto o error. Por lo tanto, una serie temporal que utiliza un modelo aditivo se puede considerar como:

$ y_t $ = $ T_t $ + $ C_t $ + $ S_t $ + $ I_t $

mientras que, un modelo multiplicativo sera:

$ y_t $ = $ T_t $ x $ C_t $ x $ S_t $ x $ I_t $

Se usara un modelo aditivo cuando las variaciones en torno a la tendencia no varen con el nivel de las series temporales, mientras que como modelo multiplicativo sera apropiado si la tendencia es proporcional al nivel de las series temporales.

### Descomposicin de STL

A veces, la tendencia y los componentes cclicos se agrupan en uno, llamado componente de tendencia-ciclo. El componente de ciclo de tendencia se puede denominar simplemente componente de "tendencia", aunque puede contener un comportamiento cclico. Por ejemplo, una descomposicin estacional de series de tiempo, conocida como **S** easonal & **T** rend **descomposicin** utilizando **L** oess o **STL**, descompone una serie de tiempo en componentes estacionales, de tendencia e irregulares utilizando loess y grafican los componentes por separado, por lo que el componente cclico (si est presente en los datos) se incluye en el grfico del componente de "tendencia".

Como su nombre indica, este modelo hace lo siguiente:

- La tendencia se elimina utilizando un modelo de regresin LOESS.
- El componente estacional se elimina mediante una regresin en componentes peridicos.
- El resto se conoce como el residual.

A continuacin, aplicaremos el modelo de descomposicin STL a nuestras series de tiempo transformadas de registro y veremos si esto elimina efectivamente la tendencia y la estacionalidad.

```r
# Funcin para la descomposicin STL de una serie de tiempo en componentes
ts.decomp <- function(df, col = 'Series Name', span = 0.13, Mult = TRUE)
{
  if(Mult) temp = log(df)  else temp = df
  spans = span * length(temp)  
  fit <- stl(temp, s.window = "periodic", t.window = spans)
  plot(fit, main = paste('Decompositon of',col,'with loess span = ', as.character(span)))
  fit$time.series
}

itca.pm25.decomp <- ts.decomp((itca.pm25.ts), col = 'Pm2.5 ITCA-FEPADE', Mult = FALSE)
itca.pm10.decomp <- ts.decomp((itca.pm10.ts), col = 'Pm10 ITCA-FEPADE', Mult = FALSE)
itca.temp.decomp <- ts.decomp((itca.temp.ts), col = 'Temperatura ITCA-FEPADE', Mult = FALSE)
str(itca.pm25.decomp)
str(itca.pm10.decomp)
str(itca.temp.decomp)

#Plot the ACF & PACF of the remainder
options(repr.pmales.extlot.width=8, repr.plot.height=6)
plot.acf(itca.pm25.decomp[, 1], is.df = FALSE)
plot.acf(itca.pm10.decomp[, 1], is.df = FALSE)
plot.acf(itca.temp.decomp[, 1], is.df = FALSE)
```

Si el proceso STL ha eliminado la tendencia y los patrones con xito, la distribucin del resto o residual debe ser i.i.d. con ACF y PACF que no muestran retrasos significativos distintos de cero. De los trazados de ACF y PACF del residuo, todava vemos algunos valores significativos de retardo no cero. As que el proceso STL no ha eliminado la tendencia o los patrones completamente.

## Modelos ARIMA para la serie residual.

Ahora que hemos investigado las propiedades bsicas de las series de tiempo y el mtodo de descomposicin de STL, investiguemos los modelos para tratar los residuos. Trabajaremos con lo que se conoce y los modelos **modelo autorregresivo integrado de promedio mvil** o **ARIMA**. En su ncleo, estos modelos son solo casos especiales de regresin lineal.

### Modelo Autoregresivo

Los valores de una serie de tiempo **autoregrisiva** o **AR** se determinan mediante una combinacin lineal de los valores pasados. En otras palabras, el modelo AR representa la correlacin serial en los valores de la serie de tiempo. Podemos escribir el valor de una serie autorregresiva o **ordenar p** o **AR (p)** series en el momento t como:

$$ x_t = \ alpha_1 x_ {t-1} + \ alpha_2 x_ {t-2} \ dots \ alpha_p x_ {t-p} + w_t $$

Un proceso AR tiene las siguientes propiedades:

- $ \ rho_0 = 1 $ siempre.
- $ p_k = \ alpha ^ k $
- Nmero de valores de PACF distintos de cero = p.

Los modelos AR son especficamente para **series temporales estacionarias**. Si la varianza no es constante o el componente de tendencia no se ha eliminado, los modelos AR no producirn resultados estadsticos.

### Modelo de media mvil

Para un modelo **de media mvil** o **MA**, el valor de la serie de tiempo en el tiempo `t` est determinado por una combinacin lineal de trminos de ruido blanco pasados. En otras palabras, el modelo de MA explica la correlacin de series en trminos de ruido. Podemos escribir el modelo MA (q) como la combinacin lineal de los ltimos trminos de ruido blanco `q` $ w_i $:

$$ x_t = w_t + \ beta_1 w_ {t-1} + \ beta_2 w_ {t-2} + \ cdots + \ beta_q w_ {t-q} $$

Un proceso de MA tiene las siguientes propiedades:

- $ \ rho_0 = 1 $ siempre. (Auto correlacin en el retraso 0)
- Nmero de $ \ rho_k distinto de cero; k \ ne 0 $ valores = q. (auto correlacin en el retraso k)

Los modelos MA son especficamente para **series de tiempo estacionarias**. Si la varianza no es constante o el componente de tendencia no se ha eliminado, los modelos MA no producirn resultados estadsticos.

### El modelo de media mvil autoregresivo

Podemos combinar los modelos AR y MA para crear un **modelo de media mvil autorregresiva** o **ARMA**. Este modelo explica la correlacin serial tanto en trminos de ruido como en valores. Podemos escribir un modelo ARMA de orden `(p, q)` como:

$$ x_t = \ alpha_1 x_ {t-1} + \ alpha_2 x_ {t-2} \ dots \ alpha_p x_ {t-p}
w_t + \ beta_1 w_ {t-1} + \ beta_2 w_ {t-2} + \ cdots + \ beta_q w_ {t-q} $$

### Modelo de media mvil integrada autoregresiva

El modelo **de media mvil integrada autorregresiva**, o **ARIMA** agrega un trmino integrador al modelo ARMA. El componente de integracin realiza la diferenciacin para modelar un componente de paseo aleatorio. El componente integrante modela una de las **partes no estacionarias** de una serie de tiempo. El modelo ARIMA est definido por rdenes p, d, q. El orden del operador de diferenciacin del trmino integrador est definido por `d`.

### Evaluacin

El paquete R `forecast` en particular, tiene herramientas para estimar automticamente los modelos ** ARIMA ** estacionales y utilizar estos modelos para hacer pronsticos futuros.

La funcin `auto.arima` del paquete` forecast` hace lo siguiente:

- Realiza una regresin progresiva progresiva a partir del orden mximo de cada coeficiente para encontrar el modelo ARIMA estacional ptimo. La opimizacin se realiza utilizando los Criterios de Informacin de Akaike (AIC) y los Criterios de Informacin Bayesiana (BIC).
- El modelo ARIMA estacional puede escribirse como ARIMA (p, d, q) (P, D, Q). Ya hemos explorado el modelo estndar de ARIMA, ARIMA (p, d, q). Los trminos estacionales adicionales modelan el comportamiento estacional y de tendencia:
- P es el orden del modelo autorregresivo estacional.
- D es el orden de un operador diferenciador o integrador, que modela la tendencia.
- Q es el orden del modelo de media mvil estacional.

Los ** Criterios de Informacin Bayesiana ** o ** BIC ** estn estrechamente relacionados con los Criterios de Informacin de Akaike. El BIC pesa el nmero de parmetros en el modelo por el registro del nmero de observaciones.

Esta es la parte interesante. Utilizaremos `auto.arima` para ajustar un modelo lineal a nuestra serie de tiempo mensual de calidad del aire para ITCA-FEPADE Sede Central. Ya tenemos las tasas medidad reales u observadas disponibles de Junio a Noviembre de 2019 en https://docs.google.com/spreadsheets/d/1oNcv1dvcbH_ygVoerA3FJ_dZn11Tzgjp3Ielky_QkRM/edit#gid=0. Podemos comparar las tasas pronosticadas con las tasas reales para ver si nuestro modelo es adecuado o no.

```r
fit.itcapm25 <- auto.arima(itca.pm25.ts)
summary(fit.itcapm25)

fit.itcapm10 <- auto.arima(itca.pm10.ts)
summary(fit.itcapm10)

fit.itcatemp <- auto.arima(itca.temp.ts)
summary(fit.itcatemp)
```

De estos resultados notamos los siguientes puntos:

- El modelo estimado es ARIMA (1,1,4) (0,0,2).
- Sigma ^ 2 de la serie original = 0.046737234933687
- sigma ^ 2 del modelo estimado = 0.001906
- Los errores estndar de los coeficientes del modelo son generalmente un orden de magnitud menor que los valores de los coeficientes del modelo. Todos los coeficientes de este modelo parecen ser significativos
- Se producen una serie de mtricas de error que incluyen:
- Error medio, ME.
- Error cuadrtico medio, RMSE.
- Error absoluto medio, MAE.
- Error de prediccin media, MPE, error de prediccin absoluta media, MAPE y error de cuadratura absoluta media, o MASE. El error de prediccin se calcula actualizando iterativamente el modelo en una ventana en expansin y midiendo el error de predecir el siguiente valor ms all de la ventana. Tenga en cuenta que el ** MAPE es bastante alto ** en comparacin con los otros errores.
- ACF1 es la correlacin del modelo residual en el primer desfase distinto de cero, o retardo 1, que no es significativo en este caso.

Usaremos el modelo ptimo estimado para hacer predicciones en el futuro.

```r
unemp.forecast <- forecast((fit.itcapm25),h=12)
summary(unemp.forecast)
plot(unemp.forecast)

unemp.forecast <- forecast((fit.itcapm10),h=12)
summary(unemp.forecast)
plot(unemp.forecast)

unemp.forecast <- forecast((fit.itcatemp),h=12)
summary(unemp.forecast)
plot(unemp.forecast)
```

La lnea azul en el grfico anterior muestra las tasas pronosticadas y el frotis gris denota los intervalos de confianza.

Notamos que el 80% y el 95% de los intervalos de confianza son casi proporcionales al valor pronosticado en s mismo y se hacen cada vez ms amplios a medida que avanza el tiempo. Esto implica que la varianza puede estar aumentando con el tiempo, o la heterocedasticidad. Eso significa que esto puede no ser adecuado para nuestros datos de calidad del aire.

A continuacin, comparemos los valores pronosticados con los valores reales (no ajustados por estacionalidad) obtenidos del sitio web en paralelo:
